# Exercise 2: Load the Pretrained Model in a New Session

```python
import torch
import sys
import importlib.util
import tiktoken

supplementary_file_path = '/content/sample_data/supplementary.py'
module_name = 'supplementary_module'

spec = importlib.util.spec_from_file_location(module_name, supplementary_file_path)
supplementary = importlib.util.module_from_spec(spec)
sys.modules[module_name] = supplementary
spec.loader.exec_module(supplementary)

GPTModel = supplementary.GPTModel
generate_text_simple = supplementary.generate_text_simple

GPT_CONFIG_124M = {
    "vocab_size": 50257,
    "context_length": 256,
    "emb_dim": 768,
    "n_heads": 12,
    "n_layers": 12,
    "drop_rate": 0.1,
    "qkv_bias": False
}

device = torch.device("cpu")

model_loaded = GPTModel(GPT_CONFIG_124M)
model_loaded.load_state_dict(torch.load("/content/sample_data/model.pth", map_location=device))
model_loaded.to(device)
model_loaded.eval()

def text_to_token_ids(text, tokenizer):
    encoded = tokenizer.encode(text, allowed_special={"<|endoftext|>"})
    encoded_tensor = torch.tensor(encoded).unsqueeze(0)
    return encoded_tensor

def token_ids_to_text(token_ids, tokenizer):
    flat = token_ids.squeeze(0)
    return tokenizer.decode(flat.tolist())

tokenizer = tiktoken.get_encoding("gpt2")

token_ids = generate_text_simple(
    model=model_loaded,
    idx=text_to_token_ids("Every effort moves you", tokenizer),
    max_new_tokens=50,
    context_size=GPT_CONFIG_124M["context_length"]
)

print("Output text:\n", token_ids_to_text(token_ids, tokenizer))
```
